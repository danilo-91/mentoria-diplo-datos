```
import pandas as pd
```

Leo el CSV primero

```
df_base = pd.read_csv('http://samplecsvs.s3.amazonaws.com/SalesJan2009.csv')
df_base.head()
```

Ahora mi idea es probar la función Quantile para dividir en 3 los precios del CSV. Primero quiero saber si el CSV tiene muchas mediciones y diferentes valores de precio, porque el Head no me da esa información.

```
df_base['Price'].unique()
```

No tiene tantos, pero algo tiene. Veamos el shape.

```
df_base.shape
```

Entonces no hay tantos valores diferentes, pero hay muchas mediciones. Veamos cómo quedan los cuartiles.

```
df_int = df_base
df_int['Price'] = df_int['Price'].str.replace(',','')
df_int = df_int.astype({'Price': int})
```

Primero tuvimos que limpiar un poco para poder trabajar bien con los números. Chequeamos con dtypes que esté todo bien y seguimos.

```
df_int.dtypes
```

Perfecto. Ahora sí, probamos el quartile.

```
df_int['Price'].quantile([.33,.66,.99])
```

```
def func_rango(row):
    if (row < df_int['Price'].quantile(.33)):
        val = 'bajo'
    elif ((row >= df_int['Price'].quantile(.33))  & (row < 2000)):
        val = 'medio'
    elif  (row >= 2000):   
        val = 'alto'
    else:
        val = 'otro'
    return val
```

```
func_rango(1200)
```
